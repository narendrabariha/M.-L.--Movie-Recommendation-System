```python
import pandas as pd
import numpy as np
import difflib   #-------------------------------------------------------------correction spelling mistake name
from sklearn.feature_extraction.text import TfidfVectorizer #---------------------text data convert to numeric number
from sklearn.metrics.pairwise import cosine_similarity #----------------------------checking similiarity 
```

```python
movie = pd.read_csv('IMDB.csv')
```

```python
movie.head(4)
```

```python
movie.shape
```

```python
movie.columns
```

```python
selected_columns =['titleType','originalTitle','genres', 'title','attributes','Description'] # separated columns here
```

```python
selected_columns
```

```python
for feature in selected_columns :                             # bar bar check karne k bajay 
    movie[feature]=movie[feature].fillna('')                # find null value & filled with fillna.'empty' by this function
```

```python
combine_feature = movie['titleType']+' '+movie['originalTitle']+' '+movie['genres']+' '+movie['title']+' '+movie['attributes']+' '+movie['Description']
```

```python
print(combine_feature)   # combined above all columns by above fucntion
```

```python
vectorizer = TfidfVectorizer()    
```

```python
vectorizer_vectors = vectorizer.fit_transform(combine_feature)      # text data is converted into numeric
```

```python
vectorizer_vectors
```

```python
similarity = cosine_similarity(vectorizer_vectors)      # how one movie is similar to others movie
```

```python
similarity
```

```python
similarity.shape
```

```python
movie_name = input('Enter your favorite movie: ')
```

```python
list_of_all_titles = movie['title'].to_list()  # .to_list()------all title movie name is converted into list
```

```python
list_of_all_titles                             # we have to see which movie is similar with other
```

```python
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles) # similar match 
                                                                                   # movie shows here by this function
```

```python
find_close_match                         # similar three movie found
```

```python
close_match = find_close_match[0]                    # index [0]
```

```python
close_match
```

```python
index_of_the_movie = movie [movie.title == close_match]['index'].values[0]  # find out 'Iron Man' movie index nuumber 
```

```python
index_of_the_movie # konse index number pe 'Iron man' movie available he & Once index no. knew then we can take  
                                      #  out information from 
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))   # enumerate work like for loop
```

```python
similarity_score  # shows similarity other movies score of 'Iron man' movie 
```

```python
sorted_similar_movie = sorted (similarity_score , key = lambda x:x [1] ,reverse= True) # arrange descending order --score
```

```python
sorted_similar_movie
```

```python
print('Movie suggested for you : \n')                      # \n---i,e new line

i = 1

for movieses in sorted_similar_movie:
    index = movieses[0] #-----------------------------------------------------movie ka '0' wale jo value he wo uska index hoga
    title_from_index = movie[movie.index == index]['title'].values[0]
    if i<30:  #----------------------------------------------------------------less '30' similar movie chahiye to 30 or can change
        print (i,'.',title_from_index)                     # indentation eror had occured
        i+=1
```

# Movie Recomendation System

```python
movie_name = input('Enter your favorite movie: ')                           # showing SIMILAR 'spider man' movie other movie
list_of_all_titles = movie['title'].to_list()
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles) 
close_match = find_close_match[0]      
index_of_the_movie = movie [movie.title == close_match]['index'].values[0]
similarity_score = list(enumerate(similarity[index_of_the_movie])) 
sorted_similar_movie = sorted (similarity_score , key = lambda x:x [1] ,reverse= True)  #--These '7' formulae took from above




print('Movie suggested for you : \n')                      # \n---i,e new line

i = 1

for movieses in sorted_similar_movie:
    index = movieses[0] #-----------------------------------------------------movie ka '0' wale jo value he wo uska index hoga
    title_from_index = movie[movie.index == index]['title'].values[0]
    if i<30:
        print (i,'.',title_from_index) 
        i+=1
```
